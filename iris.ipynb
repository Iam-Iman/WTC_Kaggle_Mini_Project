### Iris

```python
# libraries
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# reading file
df = pd.read_csv('iris.csv',index_col="Id")
df
```

```python
# viewing head
df.head()
```

```python
# Histogram 
sns.histplot(df['Petal Length (cm)']);
```

```python
# KDE plot 
sns.kdeplot(data=df['Petal Length (cm)'], shade=True);
```

```python
# 2D KDE plot
sns.jointplot(x=df['Petal Length (cm)'], y=df['Sepal Width (cm)'], kind="kde");
```

```python
# color coded plots
# Histograms for each species
sns.histplot(data=df, x='Petal Length (cm)', hue='Species')

# Add title
plt.title("Histogram of Petal Lengths, by Species");
```

```python
# KDE plots for each species
sns.kdeplot(data=df, x='Petal Length (cm)', hue='Species', shade=True)

# Add title
plt.title("Distribution of Petal Lengths, by Species");
```
